```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Pokemon.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
sns.distplot(df['HP'],color='green')
```

```python
sns.boxplot(df['HP'],color='c')
```

```python
sns.distplot(df['HP'],hist=False,rug=True,color="purple")
```

```python
sns.distplot(df['HP'],color="red",kde=False)
sns.despine(left=True)
```

```python
sns.lmplot('Speed','Sp. Atk',data=df,size=6,aspect=2,scatter_kws={'color': 'red'},line_kws={'color': 'orange'})
```

```python
sns.lmplot('Speed','Sp. Atk',col='Generation',data=df)
```

```python
sns.jointplot(df['Speed'],df['Sp. Atk'],kind='kde',size=7)
```

```python
plt.subplots(figsize=(10,5))
sns.barplot(df['Total'],df['Type 1'],palette="Reds")
```

```python
sns.boxplot(df['Type 1'],df['Total'],palette='Oranges')
plt.xticks(rotation = 90)
```

```python
plt.subplots(figsize=(14,6))
sns.swarmplot(df['Type 1'],df['Total'],size=6,palette="OrRd")
plt.xticks(rotation = 90)
```

```python
sns.lvplot(df['Type 1'], df['Total'],palette='PuRd')
plt.xticks(rotation = 90)
```

```python
plt.subplots(figsize=(10,7))
sns.stripplot(df["Total"],df["Type 1"],size=9,palette='Greys')
```

```python
plt.subplots(figsize=(12,7))
sns.violinplot(df["Type 1"],df["Total"])
plt.xticks(rotation=90)
```
